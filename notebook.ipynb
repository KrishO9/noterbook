```python
import os
from transformers import pipeline, BitsAndBytesConfig # Keep BitsAndBytesConfig for quantization
import torch
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm
import json

# --- Configuration ---
# Model to use (Unsloth quantized is good for memory)
MODEL_ID = "unsloth/llava-1.5-7b-hf-bnb-4bit"
# Or, if you want to try the official Hugging Face version with your own quantization:
# MODEL_ID = "llava-hf/llava-1.5-7b-hf"

# Paths (adjust if your script is in a subdirectory like 'scripts/')
# Assumes file_list.json is in the same directory as this script.
# If this script is in /kaggle/working/GeoSynthforLISS4/scripts/
# and file_list.json is also there, then "file_list.json" is correct.
FILE_LIST_JSON_PATH = "file_list.json"

# Where to save the output captions.
# If this script is in /kaggle/working/GeoSynthforLISS4/scripts/
# and you want to save to /kaggle/working/GeoSynthforLISS4/data/
CAPTIONS_JSON_OUTPUT_PATH = os.path.join("..", "data", "llava_generated_captions.json") # Saves to parent's data/ subdir
# Or, to save in the same directory as the script:
# CAPTIONS_JSON_OUTPUT_PATH = "llava_generated_captions.json"

# --- End Configuration ---


# Set environment variable for PyTorch memory allocator (run this early)
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

# --- Quantization Configuration (if needed or to ensure for Unsloth) ---
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16 # or torch.float16
)
# If MODEL_ID is not pre-quantized and you don't want quantization, set:
# quantization_config = None
# And adjust torch_dtype in model_kwargs accordingly (e.g., torch.float16 or None for full precision)

print(f"Initializing LLaVA captioner with model: {MODEL_ID}")
try:
    captioner = pipeline(
        "image-text-to-text", # Use this task for LLaVA-like models needing image + text prompt
        model=MODEL_ID,
        device_map="auto",    # Let accelerate distribute model across GPUs
        model_kwargs={"quantization_config": quantization_config, "torch_dtype": torch.bfloat16} if quantization_config else {"torch_dtype": torch.bfloat16}
    )
    print("LLaVA captioner initialized successfully.")
    if hasattr(captioner, 'model') and hasattr(captioner.model, 'hf_device_map'):
        print("Model device map:")
        print(captioner.model.hf_device_map)
except Exception as e:
    print(f"FATAL ERROR: Could not initialize pipeline: {e}")
    import traceback
    traceback.print_exc()
    exit()


# --- Load the list of image files to caption ---
if not os.path.exists(FILE_LIST_JSON_PATH):
    print(f"ERROR: file_list.json not found at '{FILE_LIST_JSON_PATH}'")
    print("Please ensure the script that creates file_list.json has run successfully and the path is correct.")
    exit()

try:
    with open(FILE_LIST_JSON_PATH, 'r') as f:
        files_to_caption = json.load(f)
    print(f"Loaded {len(files_to_caption)} files for captioning from {FILE_LIST_JSON_PATH}")
except Exception as e:
    print(f"Error loading {FILE_LIST_JSON_PATH}: {e}")
    exit()

# --- Prepare Output Directory and Load Existing Captions for Resuming ---
output_dir = os.path.dirname(CAPTIONS_JSON_OUTPUT_PATH)
if output_dir and not os.path.exists(output_dir): # Check if output_dir is not empty string
    os.makedirs(output_dir, exist_ok=True)
    print(f"Created output directory: {output_dir}")

generated_captions_dict = {}
if os.path.exists(CAPTIONS_JSON_OUTPUT_PATH):
    try:
        with open(CAPTIONS_JSON_OUTPUT_PATH, 'r') as f:
            generated_captions_dict = json.load(f)
        print(f"Loaded {len(generated_captions_dict)} existing captions from {CAPTIONS_JSON_OUTPUT_PATH} to resume.")
    except json.JSONDecodeError:
        print(f"Warning: {CAPTIONS_JSON_OUTPUT_PATH} is not a valid JSON. Starting fresh.")
    except Exception as e:
        print(f"Could not load existing captions from {CAPTIONS_JSON_OUTPUT_PATH}: {e}")


# --- Captioning Loop ---
print(f"\nStarting caption generation for {len(files_to_caption) - len(generated_captions_dict)} new images...")
new_captions_generated_this_run = 0

for i in tqdm(range(len(files_to_caption))):
    file_entry = files_to_caption[i]
    image_path = file_entry.get("target") # Use .get() for safety

    if not image_path:
        print(f"Warning: Entry {i} in file_list.json is missing 'target' key. Skipping.")
        continue

    if image_path in generated_captions_dict: # Check if already processed
        continue # Skip this image

    if not os.path.exists(image_path):
        print(f"Warning: Image file not found at '{image_path}'. Skipping.")
        generated_captions_dict[image_path] = "Error: Image file not found"
        continue

    try:
        image = Image.open(image_path).convert("RGB") # Ensure RGB

        # LLaVA 1.5 specific chat-like prompt structure
        # The <image> token is a placeholder that the model/processor understands
        llava_prompt_text = "Describe the contents of the image?" # The actual question
        # The full prompt structure is often handled by the processor when using chat templates,
        # but for pipelines, we might need to be more explicit or let the pipeline figure it out.
        # The "image-text-to-text" pipeline expects distinct image and text inputs.
        
        # For "image-text-to-text" pipeline, input is often a dictionary or list of dictionaries
        # The LLaVA model itself expects a prompt like "USER: <image>\nQuestion\nASSISTANT:"
        # The pipeline's processor should ideally convert this.
        
        # Let's construct the input text with the image token for LLaVA
        input_text_for_llava = f"USER: <image>\n{llava_prompt_text}\nASSISTANT:"

        # CORRECTED INPUT STRUCTURE FOR THE PIPELINE
        input_data = {"images": [image], "text": input_text_for_llava}

        result_list = captioner(
            # Pass the dictionary containing images (as a list) and text
            input_data,
            # No separate prompt kwarg here; it's part of input_data["text"]
            generate_kwargs={"max_new_tokens": 76, "do_sample": False}
        )
        
        if result_list and isinstance(result_list, list) and isinstance(result_list[0], dict) and "generated_text" in result_list[0]:
            full_generated_text = result_list[0]["generated_text"]
            
            # Extract text after the full input prompt (which LLaVA often includes in output)
            # or specifically after "ASSISTANT:" if present at the end of the input part.
            if input_text_for_llava in full_generated_text:
                 # This handles cases where the model repeats the prompt
                caption = full_generated_text.split(input_text_for_llava, 1)[-1].strip()
            elif "ASSISTANT:" in full_generated_text: # If only ASSISTANT: part is in output
                caption = full_generated_text.split("ASSISTANT:")[-1].strip()
            else: # Fallback if the structure is unexpected
                print(f"Warning: Standard LLaVA prompt markers not found in output for {os.path.basename(image_path)}. Using full output. Output: '{full_generated_text[:100]}...'")
                caption = full_generated_text.strip()
            
            generated_captions_dict[image_path] = caption
            new_captions_generated_this_run += 1
        else:
            print(f"Warning: Unexpected result structure from captioner for {os.path.basename(image_path)}: {result_list}")
            generated_captions_dict[image_path] = "Error: Unexpected captioner output structure"

    except Exception as e:
        print(f"Error captioning {os.path.basename(image_path)}: {e}")
        # import traceback # Keep commented, uncomment for deep debugging if errors persist
        # traceback.print_exc()
        generated_captions_dict[image_path] = f"Error during captioning: {str(e)}"

    # Save intermittently
    if (new_captions_generated_this_run > 0 and new_captions_generated_this_run % 10 == 0) or \
       (i + 1 == len(files_to_caption) and new_captions_generated_this_run > 0):
        try:
            with open(CAPTIONS_JSON_OUTPUT_PATH, 'w') as f_out:
                json.dump(generated_captions_dict, f_out, indent=2)
            print(f"  Saved {len(generated_captions_dict)} total captions to {CAPTIONS_JSON_OUTPUT_PATH} (processed {i+1} from file_list)")
        except Exception as e_save:
            print(f"  Error saving captions intermittently: {e_save}")

# Final save
try:
    with open(CAPTIONS_JSON_OUTPUT_PATH, 'w') as f_out:
        json.dump(generated_captions_dict, f_out, indent=2)
    print(f"\nFinal captions saved to {CAPTIONS_JSON_OUTPUT_PATH}")
    print(f"Total entries in caption file: {len(generated_captions_dict)}")
    print(f"New captions generated in this run: {new_captions_generated_this_run}")
except Exception as e_save:
    print(f"  Error saving final captions: {e_save}")

print("Script finished.")
```

```python
import json
import os

# --- Configuration ---
# Path to your existing captions.json file (output from the original LLaVA script)
CAPTIONS_INPUT_JSON_PATH = "/kaggle/working/GeoSynthforLISS4/data/llava_generated_captions.json" # Or wherever it was saved

# Base paths for constructing the "source" and "target" fields in the new JSON
# These should be the paths as they would be accessed from where your training script will run
# or the paths you want to appear in the final manifest.
# For Kaggle, if data is in /kaggle/input/
LISS4_BASE_ACCESS_PATH = "/kaggle/input/geosynthdata/liss4_patches"
OSM_BASE_ACCESS_PATH = "/kaggle/input/geosynthdata/osm_aligned_patches" # Your OSM patches

# Output filename for the new JSON
OUTPUT_PROMPT_LOCATION_JSON = "/kaggle/working/GeoSynthforLISS4/data/prompt_with_location.json"
# --- End Configuration ---

# Ensure the output directory exists
os.makedirs(os.path.dirname(OUTPUT_PROMPT_LOCATION_JSON), exist_ok=True)

if not os.path.exists(CAPTIONS_INPUT_JSON_PATH):
    print(f"ERROR: Input captions file not found at '{CAPTIONS_INPUT_JSON_PATH}'")
    exit()

try:
    with open(CAPTIONS_INPUT_JSON_PATH, 'r') as f:
        captions_data_dict = json.load(f)
    print(f"Successfully loaded {len(captions_data_dict)} entries from {CAPTIONS_INPUT_JSON_PATH}")
except Exception as e:
    print(f"Error loading {CAPTIONS_INPUT_JSON_PATH}: {e}")
    exit()

output_list = []
class_id_counter = 0

print("Transforming captions.json to prompt_with_location.json format...")

# Sort the items by key (image path) to ensure consistent ordering if that matters
# This helps if the original dictionary wasn't ordered and you want class IDs to be somewhat sequential
# based on filename order.
# Extracting the number from the filename for sorting:
# e.g., liss4_patch_00001.png -> 1
def get_id_from_path(path_str):
    try:
        return int(os.path.splitext(os.path.basename(path_str))[0].split('_')[-1])
    except:
        return float('inf') # Put items that don't match the pattern at the end

sorted_liss4_paths = sorted(captions_data_dict.keys(), key=get_id_from_path)


for liss4_full_path_from_caption_key in sorted_liss4_paths:
    caption = captions_data_dict[liss4_full_path_from_caption_key]

    if "Error:" in caption: # Skip entries that had an error during captioning
        print(f"Skipping entry for {liss4_full_path_from_caption_key} due to previous error: {caption}")
        continue

    class_id_counter += 1 # Increment for each valid entry

    # Derive the base filename (e.g., "liss4_patch_00010.png")
    liss4_base_filename = os.path.basename(liss4_full_path_from_caption_key)

    # Construct the corresponding OSM filename
    # Assumes LISS4 is "liss4_aligned_patch_XXXXX.png" and OSM is "osm_ref_patch_XXXXX.png"
    # OR if your LISS4 files are just "liss4_patch_XXXXX.png" from the start.
    # Adjust this logic if your naming convention is different.
    
    # If LISS4 filename is like "liss4_aligned_patch_00010.png"
    if "liss4_aligned_patch_" in liss4_base_filename:
        id_part = liss4_base_filename.replace("liss4_aligned_patch_", "") # Gets "00010.png"
        osm_base_filename = f"osm_ref_patch_{id_part}"
    # If LISS4 filename is just "liss4_patch_00010.png"
    elif "liss4_patch_" in liss4_base_filename:
        id_part = liss4_base_filename.replace("liss4_patch_", "") # Gets "00010.png"
        osm_base_filename = f"osm_ref_patch_{id_part}" # Assuming OSM files were named this way
    else:
        print(f"Warning: Could not derive OSM filename from LISS4 path: {liss4_full_path_from_caption_key}. Skipping this entry.")
        class_id_counter -=1 # Decrement as this pair won't be added
        continue


    # Construct the full paths for the output JSON using the defined base access paths
    # The LISS4 path in the caption key might already be the full /kaggle/input path
    # If so, we can use it directly or re-base it if necessary.
    # For consistency, let's re-base using LISS4_BASE_ACCESS_PATH and OSM_BASE_ACCESS_PATH.

    # Target path is the LISS-4 image that was captioned
    target_path_for_json = os.path.join(LISS4_BASE_ACCESS_PATH, liss4_base_filename)
    # Source path is the corresponding OSM image
    source_path_for_json = os.path.join(OSM_BASE_ACCESS_PATH, osm_base_filename)


    output_entry = {
        "source": source_path_for_json,
        "target": target_path_for_json,
        "prompt": caption, # This is the LLaVA generated caption for the LISS-4 image
        "class": class_id_counter
    }
    output_list.append(output_entry)

if output_list:
    try:
        with open(OUTPUT_PROMPT_LOCATION_JSON, 'w') as f_out:
            json.dump(output_list, f_out, indent=4) # Using indent=4 for pretty printing
        print(f"\nSuccessfully created '{OUTPUT_PROMPT_LOCATION_JSON}' with {len(output_list)} entries.")
    except Exception as e:
        print(f"Error writing {OUTPUT_PROMPT_LOCATION_JSON}: {e}")
else:
    print("No valid captions found to transform into the new format.")

print("Script finished.")
```

```python
import os
import shutil

# --- Configuration ---
PROJECT_ROOT = '/kaggle/working/GeoSynthforLISS4' # Your main project folder in Kaggle
CONTROLNET_MODELS_DIR = os.path.join(PROJECT_ROOT, 'ControlNet', 'models') # Source of the checkpoint

SOURCE_CHECKPOINT_NAME = 'control_sd21_ini.ckpt' # The file you just created
SOURCE_CHECKPOINT_PATH = os.path.join(CONTROLNET_MODELS_DIR, SOURCE_CHECKPOINT_NAME)

# New folder to create (will be inside PROJECT_ROOT)
NEW_FOLDER_NAME = 'prepared_control_checkpoint'
NEW_FOLDER_PATH = os.path.join(PROJECT_ROOT, NEW_FOLDER_NAME)

# --- Main Script ---

print(f"Project Root: {PROJECT_ROOT}")
print(f"Source Checkpoint Path: {SOURCE_CHECKPOINT_PATH}")
print(f"New folder to be created/used at: {NEW_FOLDER_PATH}")

# 1. Check if the source checkpoint exists
if not os.path.exists(SOURCE_CHECKPOINT_PATH):
    print(f"ERROR: Source checkpoint not found at {SOURCE_CHECKPOINT_PATH}")
else:
    try:
        # 2. Create the new destination folder if it doesn't exist
        # If it already exists, this script will simply copy into it.
        # If you want to ensure it's empty before copying, uncomment the shutil.rmtree lines.
        if os.path.exists(NEW_FOLDER_PATH):
            print(f"Folder '{NEW_FOLDER_PATH}' already exists. Files will be copied into it.")
            # Optional: Remove old folder for a clean copy
            # print(f"Removing existing folder '{NEW_FOLDER_PATH}' for a fresh copy.")
            # shutil.rmtree(NEW_FOLDER_PATH)
            # os.makedirs(NEW_FOLDER_PATH)
        else:
            os.makedirs(NEW_FOLDER_PATH)
            print(f"Successfully created folder: {NEW_FOLDER_PATH}")

        # 3. Copy the checkpoint file into the new folder
        destination_path_in_new_folder = os.path.join(NEW_FOLDER_PATH, SOURCE_CHECKPOINT_NAME)
        
        # Check if the destination file already exists, and overwrite if necessary
        if os.path.exists(destination_path_in_new_folder):
            print(f"Warning: Destination file '{destination_path_in_new_folder}' already exists. It will be overwritten.")
            
        shutil.copy2(SOURCE_CHECKPOINT_PATH, destination_path_in_new_folder) # copy2 preserves metadata
        print(f"Successfully copied '{SOURCE_CHECKPOINT_NAME}' to '{destination_path_in_new_folder}'")
        print(f"Size of copied file: {os.path.getsize(destination_path_in_new_folder) / (1024*1024):.2f} MB")

    except Exception as e:
        print(f"An error occurred: {e}")
        import traceback
        traceback.print_exc()

print("\nCopy script finished.")
```

```python
import pytorch_lightning as pl
from torch.utils.data import DataLoader
from data import Dataset
from ControlNet.cldm.logger import ImageLogger
from ControlNet.cldm.model import create_model, load_state_dict
from pytorch_lightning.callbacks import ModelCheckpoint
import os


# Configs
resume_path = "GeoSynthforLISS4/scripts/control_sd21_ini.ckpt"
batch_size = 4
logger_freq = 2000
learning_rate = 1e-5
sd_locked = True
only_mid_control = False


# First use cpu to load models. Pytorch Lightning will automatically move it to GPUs.
model = create_model("GeoSynthforLISS4/scripts/models/cldm_v21.yaml").cpu()
model.load_state_dict(load_state_dict(resume_path, location="cpu"))
model.learning_rate = learning_rate
model.sd_locked = sd_locked
model.only_mid_control = only_mid_control

checkpoint = ModelCheckpoint(
    dirpath=os.path.join("checkpoint", "geosynth"),
    filename="model_geosynth",
    every_n_train_steps=150,
)
# Misc
dataset = Dataset(
    prompt_path="GeoSynthforLISS4/scripts/prompt_with_locations.json",
    location_embeds_path="GeoSynthforLISS4/scripts/location_embeds.npy",
)
dataloader = DataLoader(
    dataset, num_workers=8, batch_size=batch_size, shuffle=True, persistent_workers=True
)
logger = ImageLogger(batch_frequency=logger_freq)
trainer = pl.Trainer(
    accelerator="gpu",
    strategy="ddp",
    devices=2,
    precision=32,
    max_epochs=42,
    callbacks=[logger, checkpoint],
    accumulate_grad_batches=16,
)

# Train!
trainer.fit(model, dataloader)
```
